```python
import numpy as np
import pandas as pd
```

```python
animes = pd.read_csv('anime.csv')
rating= pd.read_csv('rating.csv')
```

```python
animes.head()
```

```python
rating.head()
```

```python

animes.dropna(inplace=True)
animes.drop_duplicates(subset='name')
```

```python
animes.isnull().sum()
```

```python
animes.iloc[0].genre
```

```python

def convert(obj):
    return obj.split(', ')
```

```python
animes['genre']=animes['genre'].apply(convert)
```

```python
len(animes)
```

```python
animes['type']=animes['type'].apply(lambda x:x.split())
animes['episodes']=animes['episodes'].apply(lambda x:x.split())
animes['rating']=animes['rating'].apply(lambda x:np.array([x]))
animes.head()
```

```python
animes['tags']=animes['genre']+animes['type']+animes['episodes']
```

```python
animes.head()
```

```python
new_df=animes[['anime_id','name','rating','tags']]
```

```python
new_df
```

```python
def add(l):
    return  " ".join(str(x) for x in l)

        
```

```python
new_df['tags'] = new_df['tags'].apply(add)
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_df.rename(columns = {'name':'title'}, inplace = True)
```

```python
new_df
```

```python
def recommend(anime):
    index = new_df[new_df['title'] == anime].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Fullmetal Alchemist: Brotherhood')
```
